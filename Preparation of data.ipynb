## Loading images to a list using CV2 in gray scale and resizing

```python
# Pls be sure that the images are stored in folders named with particular class
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename),cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img,fixed_size)
        if img is not None:
            images.append(img)
    return images
```

```python
wimage = load_images_from_folder('Real_data/')
```

## Writing Images to a train folder named particular class with sequential names

```python
for i in range (0,400):
    cv2.imwrite('train/Normal/%i.jpg'%i,wimage[i])
    i= i+1
```
